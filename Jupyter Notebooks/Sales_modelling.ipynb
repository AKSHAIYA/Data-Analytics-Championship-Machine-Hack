```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor,VotingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("train_preprocessed.csv")
test = pd.read_csv("test_preprocessed.csv")
```

```python
train.head()
```

```python
train_X = train.drop(columns=['Sales'])
```

```python
train_y = train['Sales']
```

```python
train_y
```

```python
sample = test[['Sales']]
sample.to_csv("sample.csv", index = False)
```

```python
test_X = test.drop(columns=['Sales'])
```

```python
def model_fit(model, train_X, train_y, test_X,sample_sub,filename):
    model.fit(train_X, train_y)
    train_pred = np.abs(model.predict(train_X))
    print("Train RMSE : ",np.sqrt(mean_squared_error(train_y,train_pred,squared=False)))
    test_pred = model.predict(test_X)
    sample['Sales'] = np.abs(test_pred)
    sample.to_csv(filename, index = False)
    return np.sqrt(mean_squared_error(train_y,train_pred,squared=False))
```

```python

model = [LinearRegression(), Ridge(), Lasso(), ElasticNet(), KNeighborsRegressor()]
```

```python
for i in model:
    print(i)
    model_fit(i, train_X, train_y, test_X, sample, str(i)[:-2]+".csv")
```

```python
res = []
for i in range(1,10):
    model = KNeighborsRegressor(n_neighbors=i)
    print(i)
    err = model_fit(model, train_X, train_y, test_X, sample, str(i)+"KNN.csv")
    res.append(err)
```
